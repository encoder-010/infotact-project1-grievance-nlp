<h1>use our Sample Dataset (Syntetic dataset)</h1>

```python
# Creating a sample civic grievance dataset
data = {
    'complaint_id': range(1, 21),
    'complaint_text': [
        "There is no water supply in our area since 3 days",
        "Streetlight is broken near Gandhi Road for a week",
        "Roads are full of potholes causing accidents",
        "Garbage not collected in our colony for 10 days",
        "Water pipeline is leaking on MG Road",
        "Electric pole is damaged and causing danger",
        "The road near school is flooded with dirty water",
        "No electricity supply since last night, very urgent",
        "Sewage water overflowing onto the street",
        "Public park is very dirty, no cleaning done",
        "Water is very dirty and smells bad in taps",
        "Traffic signal not working at main junction",
        "Dead animals on road not removed for days",
        "Street dogs attacking people near market",
        "Building collapse risk on old bridge",
        "No drinking water in government hospital",
        "Electricity sparks from wire near playground",
        "Flood water not draining from colony roads",
        "Public toilet is broken and unusable",
        "Tree fell on road blocking traffic completely"
    ],
    'department': [
        'Water', 'Electricity', 'Roads', 'Sanitation', 'Water',
        'Electricity', 'Roads', 'Electricity', 'Sanitation', 'Sanitation',
        'Water', 'Transport', 'Sanitation', 'Animal Control', 'Roads',
        'Water', 'Electricity', 'Roads', 'Sanitation', 'Roads'
    ],
    'sentiment': [
        'Negative', 'Negative', 'Critical', 'Negative', 'Negative',
        'Critical', 'Negative', 'Critical', 'Negative', 'Negative',
        'Critical', 'Negative', 'Negative', 'Critical', 'Critical',
        'Critical', 'Critical', 'Negative', 'Negative', 'Negative'
    ]
}

df = pd.DataFrame(data)
print("Dataset created!")
print(df.shape)  # Shows rows and columns
df.head(20)      # Shows first 10 rows
```

<h1>Basic Level EDA</h1>

```python
import matplotlib.pyplot as plt

# How many complaints per department?
dept_counts = df['department'].value_counts()
print("Complaints per department:")
print(dept_counts)

# Bar chart
plt.figure(figsize=(10, 5))
dept_counts.plot(kind='bar', color='steelblue')
plt.title('Number of Complaints per Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('dept_distribution.png')  # Save chart
plt.show()
print("Chart saved!")
```

<h1>Word Cloud</h1>

```python
from wordcloud import WordCloud

# Join all complaints into one big text
all_text = ' '.join(df['complaint_text'].tolist())

# Generate word cloud
wordcloud = WordCloud(
    width=800, height=400,
    background_color='white',
    colormap='Blues'
).generate(all_text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Words in Complaints')
plt.savefig('wordcloud.png')
plt.show()
print("Word cloud saved!")
```
